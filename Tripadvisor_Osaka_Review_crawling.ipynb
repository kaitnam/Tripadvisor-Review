## 제대로된 코드#1

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from pprint import pprint
import requests

result51 = [] #2

    
driver=webdriver.Chrome('C:/Users/katenam/chromedriver.exe')
url=('https://www.tripadvisor.com/Hotel_Review-g14127623-d1237227-Reviews-Holiday_Inn_Osaka_Namba_an_IHG_hotel-Chuo_Osaka_Osaka_Prefecture_Kinki.html')
driver.get(url)

#     html_source=driver.page_source

time.sleep(3)

eng_button = driver.find_element(By.XPATH,'//*[@id="component_15"]/div/div[3]/div[1]/div[1]/div[4]/ul/li[1]/label')
eng_button.click()

time.sleep(5)


for j in tqdm(range(2,7,1)):
    html_source=driver.page_source
    soup=BeautifulSoup(html_source, 'html.parser')
    name=soup.find('div',{"class":"jvqAy"}).text
    reviews = soup.findAll('div',{"class":"YibKl MC R2 Gi z Z BB pBbQr"})
    time.sleep(2)
    

    for i in reviews:
        
        container=dict()
            #여기서 name을 가져오려고 하면 안가져와진다. 하지만 for문 위에 적으면 가져와진다. 
            # 내가봤을땐 review 섹션과 호텔 이름을 크롤링하는 부분이 너무 떨어져있어서 그런거 같다.
    #         try:
    #             name = i.find('div',{"class":"jvqAy"}).text
    #         except: 
    #             name = ''
        try:
            #태그 값을 그대로 가져오고 싶은데 i.find('div',{"class": "Hlmiy F1"}).span 치면 태그값을 가져와서 dataframe에 넣을수가 없다.
            #태그값을 찾아서 str으로 넣어주면 된다.
            rate = i.find('div',{"class": "Hlmiy F1"}).span
            rating = str(rate)[-11:-10]
        except:
            rating = ''
        try:
            cus_id = i.find('a',{"class":"ui_header_link uyyBf"}).text
        except:
            cus_id =''
        try:
            trans = i.find('div',{"class":"jngyZ dULAC"}).text[0:1]
        except:
            trans = ''

        try:
            title = i.find('a',{"class":"Qwuub"}).text
        except:
            title =''
        try:
            country = i.find('span',{"class":"default LXUOn small"}).text
        except:
            country = ''
        try:
            review = i.find('div',{"class":"fIrGe _T"}).text
        except:
            review = ''
        try:
            trip_type = i.find('span',{"class":"TDKzw _R Me"}).text[10:]
        except:
            trip_type = ''            
        try:
            votes = i.findAll('span',{"class":"yRNgz"})[1].text
        except:
            votes = ''
        try:
            date_stay = i.find('span',{"class":"teHYY _R Me S4 H3"}).text[14:]
        except:
            date_stay = ''



        container['name']=name
        container['rating'] = rating
        container['id']=cus_id
        container['country']=country
        container['title']=title
        container['review']=review
        container['type']=trip_type
        container['translation']=trans
        container['votes']=votes
        container['date of stay']=date_stay


        print(container)
        result51.append(container)     #2                
    time.sleep(2)                                      
    next_button = driver.find_element(By.CSS_SELECTOR,'#component_15 > div > div:nth-child(3) > div.MD > div > div > a:nth-child('+str(j)+')')
    next_button.click()

    time.sleep(5) 
print(result51) #2
```
